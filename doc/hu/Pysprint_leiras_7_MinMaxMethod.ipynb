A minimum-maximum módszeres kiértékelésre két különböző út van, az interaktív matplotlib ablak, illetve a manuális minimum maximum megadás.

````python
import numpy as np
import matplotlib.pyplot as plt
import pysprint as ps
````

````python
m = ps.MinMaxMethod.parse_raw('ifg.trt', 'ref.trt', 'sam.trt', skiprows=8, meta_len=6, delimiter=";", decimal=",")
m.chdomain()
m.slice(2.75, 4)

plt.figure()
m.plot()
````

### A manuális módszer
Ekkor a `MinMaxMethod.xmin` és `MinMaxMethod.xmax` paramétereket kell beállítani a minimumok és a maximumok helyeire. Mindkettő `np.ndarray` kell hogy legyen. Ehhez segédfüggvény a `MinMaxMethod.detect_peak` és a `MinMaxMethpd.detect_peak_cwt`. Ez a két függvény a `scipy.signal.find_peaks` és a `scipy.signal.find_peaks_cwt` függvényeket használja.

````python
xmax, ymax, xmin, ymin = m.detect_peak(pmax=0.5, pmin=4, threshold=0.15)

plt.figure()
m.plot()
plt.plot(xmax, ymax, "ko", label="maximumok")
plt.plot(xmin, ymin, "bo", label="minimumok")
plt.legend()
````

````python
import numbers
from pysprint.utils import find_nearest
import logging


def _split_on_SPP(a, val):
    if isinstance(val, numbers.Number):
        val, _ = find_nearest(a, val) 
        idx = np.where(a != val)[0]
        return np.split(a[idx], np.where(np.diff(idx) != 1)[0] + 1)
    elif isinstance(val, (list, np.ndarray)):
        for i, v in enumerate(val):
            if not np.any(a == v):
                value, _ = find_nearest(a, v)
                logging.info(f"{v} was replaced with {value}.")
                val[i] = value
        idx = np.in1d(a, val)
        split_at = a.searchsorted(val) - np.arange(0, np.count_nonzero(idx))
        return np.split(a[~idx], split_at)


def _build_single_phase_data(x, is_min=True, SPP_callbacks=None):
    y = np.array([])
    lastval = 0

    if SPP_callbacks is not None:
        x = _split_on_SPP(x, SPP_callbacks)
    else:
        x = (x,)
    if is_min: # TODO: the first value is slightly off..
        for index, i in enumerate(x):
            arr = np.asarray(i)
            if index % 2 == 0:
                y = np.append(y, lastval + np.pi * (2 * np.arange(0, len(arr)) + 1))
                y[-len(arr)] = lastval
            elif index % 2 == 1:
                y = np.append(y, lastval - np.pi * (2 * np.arange(0, len(arr)) + 1))
                y[-len(arr)] = lastval
            lastval = y[-1]
    else:
        for index, i in enumerate(x):
            arr = np.asarray(i)
            if index % 2 == 0:
                y = np.append(y, lastval + np.pi * 2 * np.arange(0, len(arr)))
            elif index % 2 == 1:
                y = np.append(y, lastval - np.pi * 2 * np.arange(0, len(arr)))
            lastval = y[-1]
    return np.concatenate(x), y
    
    
def _join_phase_data(
    negmin_x,
    negmin_y,
    posmin_x,
    posmin_y,
    negmax_x,
    negmax_y,
    posmax_x,
    posmax_y
):
    # we need to join the x axis first
    final_phase_x = np.array([])
    final_phase_y = np.array([])

    # join the y's the same way
    negs_x = np.insert(negmin_x, np.searchsorted(negmin_x, negmax_x), negmax_x)
    negs_y = np.insert(negmin_y, np.searchsorted(negmin_x, negmax_x), negmax_y)
    pos_x = np.insert(posmin_x, np.searchsorted(posmin_x, posmax_x), posmax_x)
    pos_y = np.insert(posmin_y, np.searchsorted(posmin_x, posmax_x), posmax_y)
    
%matplotlib notebook
x = np.arange(0, 2, 0.01)

x, y = _build_single_phase_data(x, is_min=False, SPP_callbacks=[1, 1.6])

plt.plot(x, y, 'ro')
````
